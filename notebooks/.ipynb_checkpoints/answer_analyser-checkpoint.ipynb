```python
import answeranalysis
import pandas as pd
import numpy as np
import yaml
import phik
import matplotlib.pyplot as plt
```

```python
with open('../config.yml') as f:
    config = yaml.safe_load(f)
```

```python
pd.set_option('display.max_columns', 100)
```

```python
answers = pd.read_csv(config['data_path'] +"/work.csv", encoding = "ISO-8859-1")
```

```python
answers.head()
```

```python
answers['SubmitDateTime'] = pd.to_datetime(answers['SubmitDateTime'])
```

```python
answers['SubmitDay'] = answers['SubmitDateTime'].dt.dayofweek
answers['SubmitWeek'] = answers['SubmitDateTime'].dt.week
answers['SubmitWeekDay'] = answers['SubmitWeek'].map(str) + "_" + answers['SubmitDay'].map(str)
```

```python
answers['SubmitDay'].unique()
answers['SubmitWeek'].unique()

# users_df = answers.groupby('SubmitDateTime')
```

```python
"""Only difficulty has nulls"""
answers.isnull().sum()
```

```python
answers.groupby('SubmitWeekDay').UserId.nunique()
```

```python
"""Students have 1 - 3 different subjects per day"""
answers.groupby(['SubmitWeek','SubmitDay']).Subject.nunique()
```

```python
descriptions = answers.groupby(['UserId', 'SubmitWeek', 'SubmitDay']).describe()
```

```python
"""Some questions can have 3 correct: this is with math"""
display(descriptions)
```

```python
answers_no_ts = answers.drop('SubmitDateTime', axis=1)
matrix_phik = answers_no_ts.phik_matrix()
```

```python
display(matrix_phik)
```

```python
sig_matrix = answers_no_ts.significance_matrix()
```

```python
display(sig_matrix)
```

```python
"""Quick and dirty number of correct per student per day"""
fig, ax = plt.subplots(figsize=(15,7))
user_answers = answers.groupby(['SubmitWeekDay', 'UserId'])['Correct'].sum()
user_answers.unstack().plot(ax=ax, kind='line')
```

```python
"""Difference in number answers correct per user"""
user_answers.unstack().plot(kind='kde')
```
